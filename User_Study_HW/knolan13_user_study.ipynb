##Code for analysis of the User Study

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel, ttest_ind

pd.set_option('display.max_columns', 500)
```

**There is only code here, all analysis in the corresponding word doc!**

Massaged the excel file to remove formating and add identifiers to column names.  1A, 2A... 10A for questions about tool A, 1B, 2B..10B for tool B and 1P, 2P..10P for the preference questions.  Gender and Age were also added as column labels.

Loaded to pandas:

```python
df = pd.read_excel("HWUserStudyAns_UPDATE.xls", index_col=0)
df
```

Grabbing the headers and partitioning out the the headers for each question set. (ie. all tool A columns)

```python
headers = list(df.columns.values)
A_tool = headers[1:11]
B_tool = headers [11:21]
P_quest = headers[21:31]
```

Using the list comprehension format, calculate all t-tests for each question. It takes the header arrays from above and correlates to the correct question. All are using the paired t-test.  The scipy function is documented here: http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html#scipy.stats.ttest_rel

Results are printed below.  

##A) Tool A versus Tool B, Analysis by Question	

```python
tool_by_question = [ttest_rel(df[A_tool[i]], df[B_tool[i]]) for i in range(10)]

print "\t t-statistic \t pvalue"
for i, t in enumerate(tool_by_question):
    print "Question {}: {:.3f} \t {:.3f}".format(i+1, t[0], t[1])
```

Above results show a nan {not a number}. This means the 2 results were exactly the same. As seen below, when processing the data, the underlying equation gets a divide by 0, therefore nan.  Data shown below:

```python
print df['3A'].values
print df['3B'].values
```

Filtered out data by tool into separate frames, transposed them for easiser access with pandas.  Ran the t-test for each user. Results are printed below

##B) Tool A versus Tool B, Analysis by User

```python
df_a = df[A_tool ].T
df_b = df[B_tool ].T

tool_by_user = [ttest_rel(df_a[i], df_b[i]) for i in range(1,20)]

print "\tt-statistic\tpvalue"
for i, t in enumerate(tool_by_user):
    print "User {}: {:.5f} \t {:.5f}".format(i+1, t[0], t[1])
```

## Analysis by AGE!!

First, filter data into age group categories.  Uses age range in Word doc. Also, calculating totals, by tool question for later analysis.

```python
df_old = df[df.Age > 40]
sum_row = {col: df_old[col].sum() for col in A_tool + B_tool}
old_total = pd.DataFrame(sum_row, index=["Total"])
df_old.head()
```

```python
df_middle = df[(df.Age >= 22) & (df.Age <= 40)]
sum_row = {col: df_middle[col].sum() for col in A_tool + B_tool}
middle_total = pd.DataFrame(sum_row, index=["Total"])
df_middle.head()
```

```python
df_young = df[df.Age < 22]
sum_row = {col: df_young[col].sum() for col in A_tool + B_tool}
young_total = pd.DataFrame(sum_row, index=["Total"])
df_young.head()
```

##1) Within Age Groups, Across Tools 

```python
old_t_result = ttest_rel(old_total[A_tool].loc['Total'].values, old_total[B_tool].loc['Total'].values)
print "Older age group: {:.5f} \t {:.3f}".format(old_t_result[0], old_t_result[1])

mid_t_result = ttest_rel(middle_total[A_tool].loc['Total'].values, middle_total[B_tool].loc['Total'].values)
print "Middle age group: {:.5f} \t {:.3f}".format(mid_t_result[0], mid_t_result[1])

young_t_result = ttest_rel(young_total[A_tool].loc['Total'].values, young_total[B_tool].loc['Total'].values)
print "Young age group: {:.5f} \t {:.3f}".format(young_t_result[0], young_t_result[1])
```

## 2) Within Age Groups, Across Tools, By Question 

```python
tool_question_old = [ttest_rel(df_old[A_tool[i]], df_old[B_tool[i]]) for i in range(10)]

print "\t t-statistic \t pvalue"
for i, t in enumerate(tool_question_old):
    print "Question {}: {:.3f} \t {:.3f}".format(i+1, t[0], t[1])
```

```python
tool_question_middle = [ttest_rel(df_middle[A_tool[i]], df_middle[B_tool[i]]) for i in range(10)]

print "\t t-statistic \t pvalue"
for i, t in enumerate(tool_question_middle):
    print "Question {}: {:.3f} \t {:.3f}".format(i+1, t[0], t[1])
```

```python
tool_question_young = [ttest_rel(df_young[A_tool[i]], df_young[B_tool[i]]) for i in range(10)]

print "\t t-statistic \t pvalue"
for i, t in enumerate(tool_question_young):
    print "Question {}: {:.3f} \t {:.3f}".format(i+1, t[0], t[1])
```

##3)Across Age Groups, Within Tools 
_____________________________________

Comparing:

* old with middle
* middle with young
* old with young.  

Uses the independent t test, Documented here:

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html#scipy.stats.ttest_ind

```python
age_within1a = ttest_ind(df_old.mean()[A_tool].values, df_middle.mean()[A_tool].values)
print "Older vs Middle, Tool A : {} \t {}".format(age_within1a[0], age_within1a[1])

age_within1b = ttest_ind(df_old.mean()[B_tool].values, df_middle.mean()[B_tool].values)
print "Older vs Middle, Tool B : {} \t {}".format(age_within1b[0], age_within1b[1])
```

```python
age_within1a = ttest_ind(df_young.mean()[A_tool].values, df_middle.mean()[A_tool].values)
print "Young vs Middle, Tool A : {} \t {}".format(age_within1a[0], age_within1a[1])

age_within1b = ttest_ind(df_young.mean()[B_tool].values, df_middle.mean()[B_tool].values)
print "Young vs Middle, Tool B : {} \t {}".format(age_within1b[0], age_within1b[1])
```

```python
age_within1a = ttest_ind(df_old.mean()[A_tool].values, df_young.mean()[A_tool].values)
print "Older vs Young, Tool A : {} \t {}".format(age_within1a[0], age_within1a[1])

age_within1b = ttest_ind(df_old.mean()[B_tool].values, df_young.mean()[B_tool].values)
print "Older vs Young, Tool B : {} \t {}".format(age_within1b[0], age_within1b[1])
```

##4) Across Age Groups, Within Tools, By Question 

Comparing:

* old with middle
* middle with young
* old with young

```python
age_q_a = ttest_ind(df_old[A_tool].values, df_middle[A_tool].values)
age_q_b = ttest_ind(df_old[B_tool].values, df_middle[B_tool].values)

print "Old vs Middle:"
print "Tool A"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_a[0][i], age_q_a[1][i])
print "Tool B"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_b[0][i], age_q_b[1][i])
```

```python
age_q_a = ttest_ind(df_old[A_tool].values, df_young[A_tool].values)
age_q_b = ttest_ind(df_old[B_tool].values, df_young[B_tool].values)

print "Old vs Young:"
print "Tool A"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_a[0][i], age_q_a[1][i])
print "Tool B"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_b[0][i], age_q_b[1][i])
```

```python
age_q_a = ttest_ind(df_young[A_tool].values, df_middle[A_tool].values)
age_q_b = ttest_ind(df_young[B_tool].values, df_middle[B_tool].values)

print "Young vs Middle:"
print "Tool A"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_a[0][i], age_q_a[1][i])
print "Tool B"
for i in range(10):
    print "Question {}: {}\t{}".format(i+1, age_q_b[0][i], age_q_b[1][i])
```
