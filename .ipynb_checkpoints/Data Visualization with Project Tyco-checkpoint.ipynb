##Overview of Data:
Project Tycho Level 2 Data - tables of disease counts reported by US states and cities to the Federal Health Authorities every week since 1888. 

###About Project Tycho Data

from http://www.tycho.pitt.edu/
>Project Tycho® is a project at the University of Pittsburgh to advance the availability and use of public health data for science and policy making. Currently, the Project Tycho® database includes data from all weekly notifiable disease reports for the United States dating back to 1888. These data are freely available to anybody interested. Additional U.S. and international data will be released twice yearly.

Project Tycho: http://www.tycho.pitt.edu/about.php

The data used is the Level 1 data. According to Project Tycho, this data has been curated to show a subset of years, locations and diseases.  It is the most standardized data. It contains information from 1916 to 2009, 8 disease from all 50 states and 122 cities.  This was specifically chosen because of the standardization and reduced size. The level 2 data, while more interesting was significantly larger and would have taken more clensing.  


##Data Background:

Data was retrieved from browsing data.gov, under the health section.  This leads to healthdata.gov, spnsored by U.S. Department of Health & Human Services. The particular data is from Project Tycho from the University of Pittsburg. 

Data was retrieved through the Project Tycho API. 

Data was downloaded in tab-seperated format from: http://project-tycho.s3-website-us-east-1.amazonaws.com/ProjectTycho_Level2.zip

Pertussis (Whooping Cough) has made a resurgence in the last decade

```python
from bokeh.sampledata import us_states
from math import log10

us_states = us_states.data.copy()
del us_states["HI"]
del us_states["AK"]
state_xs = [us_states[code]["lons"] for code in us_states  ]
state_ys = [us_states[code]["lats"] for code in us_states ]

wc = df[(df.disease == 'PERTUSSIS') & (df.year >= 2009)]

wc_1 = pd.pivot_table(wc, index=['state'], values=['number'], aggfunc=np.sum, fill_value=0)
colors = ["#EFE1EA" ,"#EECAD5", "#EDB4C1", "#EC9DAD", "#EB8799", "#EA7085", "#E95A70", 
          "#E8435C", "#E72D48", "#E72D48", "#E61634"]
          

state_color = []
for state in wc_1.iterrows():
    if state[0] in ["AK", "HI"]:
        continue
    rate = state[1]['number']
    idx = min(rate/100, 10)
    state_color.append(colors[idx])

bk.figure()
bk.patches(state_xs, state_ys, fill_color=state_color, fill_alpha=0.7, 
           tools="pan,wheel_zoom,box_zoom,reset,hover,previewsave",
           line_color="black", line_width=0.5, 
           title="US Pertussis Rates 2009-2011")

hover = bk.curplot().select(dict(type=HoverTool))
hover.tooltips = OrderedDict([
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("fill color", "$color[hex, swatch]:fill_color"),
])

bk.show()
    
```
